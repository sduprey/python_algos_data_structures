# Pivot Tables

```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(
    {"class":list("ABC")*4,
     "lesson":["math","stat"]*6,
     "sex":list("MFMM")*3,
     "sibling":[1,2,3]*4,
     "score":np.arange(40,100,5)})
```

```python
df
```

```python
df.groupby("lesson")["score"].mean()
```

```python
df.groupby(
    ["lesson",
     "class"])[
    "score"].aggregate("mean").unstack()
```

```python
df.pivot_table(
    "score",
    index="lesson",
    columns="class")
```

```python
df.pivot_table(
    ["sibling","score"],
    index=["class","lesson"],
    columns="sex")
```

```python
df.pivot_table(
    ["sibling","score"],
    index=["class","lesson"],
    columns="sex",margins=True)
```

```python
df.pivot_table(
    ["sibling","score"],
    index=["class","lesson"],
    columns="sex",fill_value=0)
```

## Multi-level pivot tables

```python
sib=pd.cut(df["sibling"],[0,2,3])
```

```python
df.pivot_table("score",
               ["lesson",sib],
               "class",fill_value=0)
```

```python
df.pivot_table(
    "score",
    index="lesson",
    columns="class")
```

```python
df.pivot_table(
    index="lesson",
    columns="class",
    aggfunc="sum")
```

```python
df.pivot_table(
    index="lesson",
    columns="class",
    aggfunc={"sibling":"max",
             "score":"sum"})
```

## Cross-Tabulations: Crosstab

```python
pd.crosstab(df.sibling,df.lesson)
```

```python
pd.crosstab([df.sibling, df.lesson], df.sex)
```
